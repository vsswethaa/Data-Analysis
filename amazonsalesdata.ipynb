```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/Monthly Sales.csv'
data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Sales Per Month'], marker='o')

# Customize the plot
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Per Month')
plt.grid(True)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the data
file_path_funnel = '/content/Top Category by Sales.csv'
data_funnel = pd.read_csv(file_path_funnel)

# Sort data for better visualization
data_funnel = data_funnel.sort_values('Most Sold Category', ascending=False)

# Create a funnel chart
fig = go.Figure(go.Funnel(
    y = data_funnel['Category'],
    x = data_funnel['Most Sold Category'],
    textinfo = "value+percent initial"
))

# Customize the layout
fig.update_layout(
    title="Top Categories by Sales",
    funnelmode="stack",
    width=800,
    height=500
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path_pie = '/content/Total Quantity by Category.csv'
data_pie = pd.read_csv(file_path_pie)

# Plot the pie chart
plt.figure(figsize=(5, 5))
plt.pie(data_pie['Sum of Qty'], labels=data_pie['Category'], autopct='%1.1f%%', startangle=140)

# Customize the plot
plt.title('Total Quantity by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Load the new CSV file to check its structure
file_path_new = '/content/Total Quantity by Size.csv'
data_new = pd.read_csv(file_path_new)

# Display the data to understand its structure
data_new.head()

# Data
sizes = data_new['Size']
quantities = data_new['Sum of Qty']

# Create a pie chart with adjusted figure size
plt.figure(figsize=(6, 6))
plt.pie(quantities, labels=sizes, autopct=lambda p: '{:.0f}'.format(p * sum(quantities) / 100),
        startangle=140, wedgeprops=dict(width=0.3))

# Draw a circle in the center to make it a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.title('Total Quantity by Size')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Load the new CSV file to check its structure
file_path_new = '/content/Count of Orders by Status.csv'
data_new = pd.read_csv(file_path_new)

# Display the data to understand its structure
data_new.head()

# Data
statuses = data_new['Status']
order_counts = data_new['Count of Order ID']

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(statuses, order_counts, color=['red', 'blue', 'orange', 'green'])

# Add labels and title
plt.xlabel('Order Status')
plt.ylabel('Count of Orders')
plt.title('Count of Orders by Status')

# Display the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = '/content/Order and Quantity by MonthCategory.csv'
data = pd.read_csv(file_path)

# Convert the 'Month' column to a categorical type with a specific order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
data['Month'] = pd.Categorical(data['Month'], categories=month_order, ordered=True)

# Create a FacetGrid for small multiples
g = sns.FacetGrid(data, col="Category", col_wrap=4, height=4, aspect=1.5, sharey=False)

# Map the data to the grid
g.map(sns.lineplot, "Month", "Count of Qty")

# Adjust labels and titles
g.set_axis_labels("Month", "Count of Qty")
g.set_titles("{col_name} Category")
g.set_xticklabels(rotation=45)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Quantity Count by Month and Category', fontsize=16)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
file_path = '/content/Order Count  and Average Quantity by Month.csv'
 # Replace with your file path
data = pd.read_csv(file_path)

# Set the figure size
plt.figure(figsize=(10, 6))

# Define the positions for the bars
bar_width = 0.35
index = np.arange(len(data['Month']))

# Plot the clustered columns
plt.bar(index, data['Count of Order ID'], bar_width, label='Count of Order ID')
plt.bar(index + bar_width, data['Average of Qty'], bar_width, label='Average of Qty')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Count of Order ID and Average of Qty by Month')
plt.xticks(index + bar_width / 2, data['Month'])
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
